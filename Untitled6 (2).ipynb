```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("canadian_immegration_data.csv")
```

```python
df.info()  
df.head()
```

```python
# 1. Analyze the immigration data for Yemen, Zimbabwe, and Zambia by comparing their mean, median, and mode to identify the skewness of each distribution.



countries_interest = ['Yemen', 'Zimbabwe', 'Zambia']  
country_data = df[df['Country'].isin(countries_interest)]  

```

```python
numeric_data = country_data.select_dtypes(include=[np.number])
stats = numeric_data.describe().T[['mean', '50%', 'std']]  
stats.rename(columns={'50%': 'median'}, inplace=True) 


modes = numeric_data.mode().iloc[0]  
stats['mode'] = modes  

stats['skewness'] = numeric_data.skew()  

print(stats) 
```

```python
# 2. Identify the top 10 countries with the highest average immigration over all years. What percentage of total immigration do these top 10 countries contribute?

df['Average_Immegration'] = df.select_dtypes(include=[np.number]).mean(axis=1)  
top_10_countries = df.nlargest(10, 'Average_Immegration') 

total = df.select_dtypes(include=[np.number]).sum().sum()  
top_10_contries = (top_10_countries['Average_Immegration'].sum() / total) * 100  

print("Top 10 Countries with HighestImmegration:")
print(top_10_countries[['Country', 'Average_Immegration']])  
print(f"Percentage Top 10 Countries: {top_10_contries:.2f}%")
```

```python
# 3. Create a box plot of immigration data for the top 5 countries with the highest median immigration. Identify any outliers and describe the skewness of their distributions.


df['Median_Immegration'] = df.select_dtypes(include=[np.number]).median(axis=1)  
top_5_countries = df.sort_values(by='Median_Immegration', ascending=False).head(5) 


plt.figure(figsize=(15, 8))  
sns.boxplot(data=top_5_countries.select_dtypes(include=[np.number]).T) 
plt.xticks(rotation=45, fontsize=12) 
plt.yticks(fontsize=12)  
plt.title("Box Plot of Immegration Data for Top 5 Countries", fontsize=14) 
plt.xlabel("Year", fontsize=12) 
plt.ylabel("Number of Immegrants", fontsize=12)  
plt.show()
```

```python
for country in top_5_countries['Country']: 
    country_data = df[df['Country'] == country].select_dtypes(include=[np.number])  
    skewness = country_data.skew().mean()  
    print(f"Country: {country}, Skewness: {skewness:.2f}")
```

```python
developed_count = df[df['DevName'] == 'Developed regions'].shape[0]
developing_count = df[df['DevName'] == 'Developing regions'].shape[0]

total_countries = df.shape[0]

probability_developed = developed_count / total_countries
print(f"Probability of selecting a developed country: {probability_developed:.2f}")

labels = ['Developed', 'Developing']
sizes = [developed_count, developing_count]


plt.figure(figsize=(9, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal') 
plt.title('Proportion of Developed v/s Developing Countries')
plt.show()
```

```python
# 6. Create a new column showing the percentage contribution of each countryâ€™s immigration to the total immigration count.

total = df.select_dtypes(include=[np.number]).sum(axis=1).sum()  
df['Percentage'] = (df.select_dtypes(include=[np.number]).sum(axis=1) / total) * 100  


print(df[['Country', 'Percentage']]) 
```

```python
# Visualize the immigration trends for different continents using a line graph. Which continent had the most significant growth in immigration?
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sns 
df = pd.read_csv("canadian_immegration_data.csv")
if 'Continent' in df.columns:
    print("Unique values':", df['Continent'].unique())  
    
    
    trends = df.groupby('Continent').sum(numeric_only=True).T  
    
  
    plt.figure(figsize=(12, 6)) 
    for continent in trends.columns:
        plt.plot(trends.index, trends[continent], label=continent)  
    
    plt.title("Immegration Trends") 
    plt.xlabel("Year") 
    plt.ylabel("Number of Immegrants") 
    plt.xticks(rotation=65)  
    plt.legend(title="Continent")  
    plt.grid(True) 
    plt.show()  
    
    
    continent_growth = trends.iloc[-1] - trends.iloc[0]  
    most_growth_continent = continent_growth.idxmax() 
    print(f"Continent with the most growth: {most_growth_continent}")
else:
    print("Column 'Continent' not found. ")
```
